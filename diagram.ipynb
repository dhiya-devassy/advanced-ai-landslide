```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_nodes_from([
    "Data Sources", "Weather & Geospatial Data", "Seismic Data", 
    "Machine Learning Models", "Random Forest", "CNN", "SVM", 
    "Real-Time Sensors", "Emergency Notifications", 
    "Safety Container (AI)", "Health Monitoring", "Survivor Detection",
    "Authorities", "Rescue Teams", "Occupants"
])

# Add edges
edges = [
    ("Data Sources", "Weather & Geospatial Data"),
    ("Data Sources", "Seismic Data"),
    ("Weather & Geospatial Data", "Random Forest"),
    ("Seismic Data", "CNN"),
    ("Seismic Data", "SVM"),
    ("Machine Learning Models", "Random Forest"),
    ("Machine Learning Models", "CNN"),
    ("Machine Learning Models", "SVM"),
    ("Real-Time Sensors", "CNN"),
    ("CNN", "Emergency Notifications"),
    ("CNN", "Safety Container (AI)"),
    ("Safety Container (AI)", "Health Monitoring"),
    ("Safety Container (AI)", "Survivor Detection"),
    ("Emergency Notifications", "Authorities"),
    ("Safety Container (AI)", "Rescue Teams"),
    ("Safety Container (AI)", "Occupants")
]
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(14, 10))
pos = nx.spring_layout(G, seed=42)  # Layout for the graph
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='skyblue')
nx.draw_networkx_edges(G, pos, arrowstyle='-|>', arrowsize=20)
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black')

plt.title("Technical Architecture of Advanced Landslide Response System", fontsize=16)
plt.show()
```
